```python
from bs4 import BeautifulSoup
import urllib

url1 = 'http://export.arxiv.org/api/query?search_query=all:supernova&start=0&max_results=10000'
data = urllib.request.urlopen(url1).read()

soup = BeautifulSoup(data)
```

```python
abstract = soup.find_all('summary')
len(abstract)
```

```python
# Write text to file

n = 2500
txtfile = open('abstracts_{}.txt'.format(n), 'w')

for i, text in enumerate(abstract[:n]):
    
    print('Writing abstracts to file ... {0:.2f}%'.format((i+1)*100/len(abstract[:n])), end='\r')
    
    text = text.text.strip()
    text = text.replace('\n', ' ')
    #print(text.replace('\n', ' '))
    txtfile.write(text)
    txtfile.write('\n')

txtfile.close()
```
